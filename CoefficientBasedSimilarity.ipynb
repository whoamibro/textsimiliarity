```python
import scipy.stats as sc
import csv
from konlpy.tag import Twitter
twitter = Twitter()
```

```python
dataset = ['쿠폰사용은 어떻게 하나요?', 
           '쿠폰은 어디서 확인하나요?', 
           '쿠폰 확인 어떻게 하나요?',
           '제가 받은 쿠폰 어디있나요?']



def tokenize(sentence):
    return [''.join(m) for m in twitter.morphs(sentence)]
```

```python
print(tokenize(dataset[0]))
```

```python
tokenizedList = [(tokenize(sentence)) for sentence in dataset]
```

```python
print(tokenizedList)
```

```python
wordList = []
for tklist in tokenizedList:
    wordList.extend(tklist)
print(wordList)
wordList = list(set(wordList))
print(wordList)
type(wordList)
```

```python
testlist = []
for ohsen in dataset:
    index = 1;
    testsen = []
    for word in wordList:
        for token in tokenize(ohsen):
            if word == token:
                testsen.append(1)
            else :
                pass
        if len(testsen) < index:
            testsen.append(0)
        index = index + 1 
    testlist.append(testsen)
print(testlist)

print(sc.stats.pearsonr(testlist[0], testlist[1]))
print(sc.stats.pearsonr(testlist[1], testlist[2]))
print(sc.stats.pearsonr(testlist[2], testlist[3]))
```

```python
import os
print (os.getcwd())
f = open('./PycharmProjects/Textsimilarity/training_data.csv', 'r')
datafile = csv.reader(f)
dataset = []

for data in datafile:
    dataset.append(data)
f.close()

morphemes = [] #list for filling all the morphemes
temp_morphemes = [["" for col in range(10)] for row in range(100)] # morphemes analysis of training data

print(dataset)
print(temp_morphemes)

```

```python
for index in range(0,99):
    temp = twitter.morphs(dataset[index][0])
    temp_morphemes[index] = temp
    morphemes.extend(temp)

morphemes = list(set(morphemes))
print(temp_morphemes)
print('\n\n')
print(morphemes)
```

```python
cal_list = [[0 for col in range(len(morphemes))] for row in range(100)]


for i in range(0, 99):
    for j in range(0, len(morphemes)):
        for k in range(0, len(temp_morphemes[i])):
            if morphemes[j] == temp_morphemes[i][k]:
                cal_list[i][j] = 1
                
print(cal_list)
```
